```python
import numpy as np
# importing seaborn which is a python library for data visualization. This will help in plotting pairwise relationship.
import seaborn as sea_q1
# importing pandas library for converting sklearn dataset to dataframe
import pandas as pd     
#importing matplotlib library for plotting 2-D plots.
import matplotlib.pyplot as plot_q1          
#loading load_iris dataset of flowers.
from sklearn.datasets import load_iris             
```

```python
#storing the sklearn dataset in a variable for ease of use.
iris_dataset=load_iris() 
conv_data_in_dataframe= pd.DataFrame(np.column_stack((iris_dataset.data, iris_dataset.target)),columns = iris_dataset.feature_names+['target']) 
```

<h5> As I am making the pairwise relationship using seaborn, So I need the dataframe as a parameter to pass.So converting to dataframe. Not needed if plotting is being done by other libraries.
np.column_stack :: concatinating the numpy arrays of data(containing SL,SW,PL,PW) and target(0..,1..,2..)
columns:: To assign the names for dataframe columns. Pandas is used to convert the sklearn dataset into a DF.</h5>

```python
# Pairplot is used to make a pairwise relationship. Dataframe is passed as data for seaborn.pairplot which considers df 
#as collection of rows and columns. diag_kind=hist because it was taking it as kde plot.
#Target is not passed in the vars, otherwise it will make a plot of target vs target too.Setting hue for 
#colors as per target. height is set as 2.8 and passed dark to palette for darker colors in the plot.
plotting=sea_q1.pairplot(conv_data_in_dataframe,diag_kind='kde',vars=["sepal length (cm)", "sepal width (cm)", "petal length (cm)","petal width (cm)"],hue="target",height=2.8,palette="dark"); 
#Removing the legend explicitly because it was overlapping with the graphs and I had to use zipped for loop for 
#performing the label-name change. As target is 0..,1..,2.so for changing that to target-names, I had to use a 
#zip for loop with labels and title. 
plotting._legend.remove() 
#plotting the legend separately with loc set as "best",passing list of required labels and title 
#and setting bbox_anchor with the parameters x and y.
plot_q1.legend(loc='best', title='Iris-Species',labels=['Iris-setosa', 'Iris-versicolor','Iris-virginica'], ncol=1,bbox_to_anchor=(1.0,1.0)) 
```

<h2>--OBSERVATIONAL ANALYSIS--</h2>


<h3>Petal length and petal width: Best separation between classes</h3>

<p>* These two are the best features which separate all the 3 classes without showing any overlapping in the classification points because these two features distingushes the classes very well when we look at the dataset.</p>

<h3>Sepal length and sepal width: Bad separation between classes</h3>

<p>* Sepal length and sepal width are the least uselful features for finding the separation between the classes as these make the wrong classifications leading to the low accuracy rate.</p>

<h6> It can be clearly analyzed from the graphs that setosa can be separated very easily whereas the versicolor and virginica are overlapping at some points.</h6>
